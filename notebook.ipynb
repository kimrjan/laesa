```python
from matplotlib import pyplot as plt
import numpy as np
from laesa import Index, Distribution
```

Index pro většinu testování výkonu. V indexu je `10 000` hodnot z rozmezí 0 až 1. Pivoty jsou určeny jak centra shlůků získaná pomocí algoritmu kmeans.

```python
index = Index(2)
index.add_data(10000, [Distribution('uniform', low=0, high=1), Distribution('uniform', low=0, high=1)])
index.calc_index(10, 'kmeans')
```

```python
result = index.range_query([0.5, 0.5], 0.05)
print(result)
```

```python
result = index.knn_query([0.5, 0.5], 50)
print(result)
```

Získání statistik podle počtu pivotů.

```python
iter_cnt = 20
pivot_cnt = 20


def get_avg_stats_knn(index_, k, dim=2, use_index=True):
    stats = np.empty((iter_cnt, 4))
    for row in range(iter_cnt):
        stats[row] = index_.knn_query(np.random.default_rng().uniform(0, 1, dim), k, use_index=use_index, quiet=True,
                                      return_stats=True)

    return np.array(stats).mean(axis=0)


def get_summary_knn(k):
    summary = np.empty((pivot_cnt, 4))
    summary[0] = get_avg_stats_knn(index, k, use_index=False)
    for i in range(1, pivot_cnt):
        index.calc_index(i, 'kmeans')
        summary[i] = get_avg_stats_knn(index, k)

    return summary
```

```python
pivots_avg_knn_stats_200 = get_summary_knn(200)
pivots_avg_knn_stats_1000 = get_summary_knn(1000)
pivots_avg_knn_stats_2000 = get_summary_knn(2000)
```

```python
def get_avg_stats_range(index_, max_range, dim=2, use_index=True):
    stats = np.empty((iter_cnt, 4))
    for row in range(iter_cnt):
        stats[row] = index_.range_query(np.random.default_rng().uniform(0, 1, dim), max_range, use_index=use_index,
                                        quiet=True, return_stats=True)

    return np.array(stats).mean(axis=0)


def get_summary_range(max_range):
    summary = np.empty((pivot_cnt, 4))
    summary[0] = get_avg_stats_range(index, max_range, use_index=False)
    for i in range(1, pivot_cnt):
        index.calc_index(i, 'kmeans')
        summary[i] = get_avg_stats_range(index, max_range)

    return summary
```

```python
pivots_avg_range_stats_005 = get_summary_range(0.05)
pivots_avg_range_stats_010 = get_summary_range(0.1)
pivots_avg_range_stats_025 = get_summary_range(0.25)
```

```python
def plot_avg_stats(position, data, title):
    plt.subplot(position)
    plt.plot(data[:, 0], 'o-r', color='blue')
    plt.title(title)
    plt.xlabel('Pivot count')
    plt.ylabel('Time (microseconds)')
    plt.subplot(position + 1)
    plt.plot(data[:, 1], 'o-r', color='red')
    plt.title(title)
    plt.xlabel('Pivot count')
    plt.ylabel('Distance calculation called')
```

```python
plt.figure(figsize=(16, 20))

plot_avg_stats(321, pivots_avg_knn_stats_200, 'knn query (k=200)')
plot_avg_stats(323, pivots_avg_knn_stats_1000, 'knn query (k=1000)')
plot_avg_stats(325, pivots_avg_knn_stats_2000, 'knn query (k=2000)')

plt.show()
```

```python
plt.figure(figsize=(16, 20))

plot_avg_stats(321, pivots_avg_range_stats_005, 'range query (range=0.05)')
plot_avg_stats(323, pivots_avg_range_stats_010, 'range query (range=0.10)')
plot_avg_stats(325, pivots_avg_range_stats_025, 'range query (range=0.25)')

plt.show()
```

```python
def get_summary_without_index_range(max_range, use_index):
    max_iter = 14
    summary = np.empty((max_iter, 5))

    for idx in range(max_iter):
        index_sizes = Index()
        index_sizes.add_data(16 * np.power(2, idx))
        index_sizes.calc_index(10, 'kmeans')
        summary[idx, 0] = 16 * np.power(2, idx)
        summary[idx, 1:] = get_avg_stats_range(index_sizes, max_range, use_index=use_index)

    return summary


def get_summary_without_index_knn(k, use_index):
    max_iter = 14
    summary = np.empty((max_iter, 5))

    for idx in range(max_iter):
        index_sizes = Index()
        index_sizes.add_data(16 * np.power(2, idx))
        index_sizes.calc_index(10, 'kmeans')
        summary[idx, 0] = 16 * np.power(2, idx)
        summary[idx, 1:] = get_avg_stats_knn(index_sizes, k, use_index=use_index)

    return summary
```

```python
def plot_without_index_stats(position, with_index, without_index, title):
    plt.subplot(position)
    plt.plot(with_index[:, 0], with_index[:, 1], 'o-r', color='blue', label='with index')
    plt.title(f'{title}')
    plt.xlabel('Dataset size')
    plt.ylabel('Time (microseconds)')
    plt.plot(without_index[:, 0], without_index[:, 1], 'o-r', color='red', label='without index')
    plt.legend()
```

```python
avg_knn_stats_with_index_200 = get_summary_without_index_knn(200, True)
avg_knn_stats_with_index_1000 = get_summary_without_index_knn(1000, True)
avg_knn_stats_with_index_2000 = get_summary_without_index_knn(2000, True)

avg_knn_stats_without_index_200 = get_summary_without_index_knn(200, False)
avg_knn_stats_without_index_1000 = get_summary_without_index_knn(1000, False)
avg_knn_stats_without_index_2000 = get_summary_without_index_knn(2000, False)
```

```python
avg_range_stats_with_index_005 = get_summary_without_index_range(0.05, True)
avg_range_stats_with_index_010 = get_summary_without_index_range(0.1, True)
avg_range_stats_with_index_025 = get_summary_without_index_range(0.25, True)

avg_range_stats_without_index_005 = get_summary_without_index_range(0.05, False)
avg_range_stats_without_index_010 = get_summary_without_index_range(0.1, False)
avg_range_stats_without_index_025 = get_summary_without_index_range(0.25, False)
```

```python
plt.figure(figsize=(16, 20))

plot_without_index_stats(321, avg_knn_stats_with_index_200, avg_knn_stats_without_index_200, 'knn query (k=200)')
plot_without_index_stats(323, avg_knn_stats_with_index_1000, avg_knn_stats_without_index_1000, 'knn query (k=1000)')
plot_without_index_stats(325, avg_knn_stats_with_index_2000, avg_knn_stats_without_index_2000, 'knn query (k=2000)')

plot_without_index_stats(322, avg_range_stats_with_index_005, avg_range_stats_without_index_005,
                         'range query (range=0.05)')
plot_without_index_stats(324, avg_range_stats_with_index_010, avg_range_stats_without_index_010,
                         'range query (range=0.10)')
plot_without_index_stats(326, avg_range_stats_with_index_025, avg_range_stats_without_index_025,
                         'range query (range=0.25)')

plt.show()
```

```python
dimensions = [2, 3, 5, 10, 20, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150]


def get_summary_dimensions_range(max_range):
    summary = np.empty((len(dimensions), 5))

    for idx, dim in enumerate(dimensions):
        index_dimensions = Index(dim)
        index_dimensions.add_data(10000, [Distribution('uniform', low=0, high=1) for _ in range(dim)])
        index_dimensions.calc_index(10, 'kmeans')
        summary[idx, 0] = dim
        summary[idx, 1:] = get_avg_stats_range(index_dimensions, max_range, dim)

    return summary
```

```python
dimensions_range_stats = get_summary_dimensions_range(0.2)
```

```python
plt.figure(figsize=(20, 8))

plt.subplot(121)
plt.plot(dimensions_range_stats[:, 0], dimensions_range_stats[:, 1], 'o-r', color='blue')
plt.title('range query (range=0.2)')
plt.xlabel('Dimensions')
plt.ylabel('Time microseconds')

plt.subplot(122)
plt.plot(dimensions_range_stats[:, 0], dimensions_range_stats[:, 2], 'o-r', color='red')
plt.title('range query (range=0.2)')
plt.xlabel('Dimensions')
plt.ylabel('Distance calculation called')

plt.show()
```
